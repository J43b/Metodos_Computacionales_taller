# Estudio de las geodésicas en el espacio-tiempo de Schwarzschild 

### Planteamiento del problema:

Deducción de las geodésicas planetarias en el campo gravitatorio de un objeto masivo esféricamente simétrico a partir de la métrica de Schwarzschild: un análisis comparativo con la teoría Newtoniana.

### Teoria:

##### Teoria Newtoniana

La energía total de un cuerpo de masa $m_{0}$ que se desplaza en un campo gravitatorio newtoniano creado por una estrella esféricamente simétrica de masa $m$ puede ser calculada utilizando el potencial gravitatorio $V(r) = - \frac{Gm}{r}$. Esta energía total se define como:

$$E_N =  \frac{m_0 \dot{r}^2}{2} + \frac{J_N^2}{2m_0r^2} - \frac{Gmm_0}{r} \tag{1.1}$$

Donde $J_{N}$ es el momento angular y se expresa como $J_{N} = m_{0}r^2\dot{\phi} = $ cte. Haciendo $u = \frac{m}{r}$ y derivando respecto a $\varphi$ resulta

\begin{equation*}
\frac{du}{d\varphi} = - \frac{M}{r^2} \frac{dr}{ds} \frac{ds}{d\varphi} = - \frac{M}{r^2} \frac{\frac{dr}{ds}}{\frac{ds}{d\varphi}} = - \frac{M}{J} \frac{dr}{ds}.
\end{equation*}

Reemplazando en (1.1)

\begin{equation*}
\frac{du}{d\varphi}^2= 2\frac{m_{0}m^2}{J_{N}^2}(E_{N}- \frac{J_{N}^2}{2m_{0}r^2} + \frac{Gmm_0}{r})
 \\= \frac{2m_{0}mE_{N}}{J_{N}^2}-u^2+\frac{2Gm_{0}^2m^2}{J_{N}^2}u
\end{equation*}

Diferenciando con respecto a $\varphi$ y dividiendo por $2\frac{du}{d\varphi}$ se encuentra la ecuación newtoniana del movimiento:

\begin{equation*}\tag{1.2}
\frac{d^2u}{d\phi^2} = -u+ \mathring{u}, \quad \mathring{u}=\frac{Gm^2m_{0}^2}{J_{N}^2}
\end{equation*}

La solucion analitica de (1.2) esta dada por:

\begin{equation*}\tag{1.3}
\frac{m}{r}\equiv u= \mathring{u}(1+e\cos{(\phi-\phi_{0})})
\end{equation*}

Donde $e$ es la excentricidad.


##### Teoría Relativista

Para nuestro estudio relativista del movimiento orbital del planeta Mercurio en el campo gravitacional del Sol, el cual ha desempeñado un papel crucial como prueba de validez de la Relatividad General, vamos a utilizar la métrica de Schwarzschild definida de la siguiente manera:

\begin{equation}\tag{2.1}
    ds^2 = - \left( 1 - \frac{2M}{r} \right) dt^2 + \frac{dr^2}{1 - \frac{2M}{r}} + r^2 \left( d\theta ^2 + \sin^2(\theta)d\varphi ^2 \right) 
\end{equation}

En esta métrica, $ ds^2 $ representa el elemento de intervalo espacio-temporal, donde $ M $ es la masa del objeto central (el Sol en este caso), $ r $ es la distancia radial, $ t $ es el tiempo, $ \theta $ es el ángulo polar y $ \varphi $ es el ángulo azimutal.

Esta formulación nos permite describir y analizar el movimiento de Mercurio teniendo en cuenta los efectos relativistas, brindando una representación precisa de cómo la presencia de masa afecta la geometría del espacio-tiempo en su entorno gravitacional.


Para encontrar la ecuación de movimiento calculamos el lagrangiano para la geodésicas de la métrica (2.1), parametrizadas por un parámetro $s$:

\begin{equation*}
L =  \frac{1}{2}\left( - \left(1 - \frac{2M}{r}\right) \left(\frac{dt}{ds}\right)^2 + \frac{1}{1 - \frac{2M}{r}} \left(\frac{dr}{ds}\right)^2 + r^2 \left( \left(\frac{d\theta}{ds}\right)^2 + \sin^2(\theta)\left(\frac{d\varphi}{ds}\right)^2 \right) \right),
\end{equation*} 

La ecuación de Euler-Lagrange para la función de coordenadas $s \longmapsto t(s)$ es la siguiente 

\begin{align*}
\frac{d}{ds} \left(r^2\left(\frac{dt}{ds}\right)\right) &= r^2 \sin(\theta)\cos(\theta)\left(\frac{d\varphi}{ds}\right)^2 ; \\
\frac{d}{ds} \left(r^2 \sin^2(\theta) \left(\frac{d\varphi}{ds}\right)\right) &= 0 ;\\
\frac{d}{ds} \left[\left(1 - \frac{2M}{r}\right) \left(\frac{dt}{ds}\right)\right] &= 0 ;

\end{align*}


No hemos escrito la ecuación para $\left(\frac{d^2r}{ds^2}\right)$. En lugar de eso es más conveniente dividir la Ec. (2.1) por $-ds$:

\begin{equation}\tag{2.2}
    1 = \left( 1 - \frac{2M}{r}\right) \left(\frac{dt}{ds}\right)^2 - \left(1 -\frac{2M}{r}\right)^{-1} \left(\frac{dr}{ds}\right)^2- r^2 \left( \left(\frac{d\theta}{ds}\right) ^2 + \sin^2(\theta)\left(\frac{d\varphi}{ds}\right)^2 \right) 
\end{equation}

ahora incluso en la métrica completamente relativista de la solución de Schwarzschild todas las órbitas estarán en planos a través del origen, ya que la simetría esférica nos permite elegir como nuestra condición inicial

$$\theta = \frac{\pi}{2}; \qquad \frac{d\theta}{ds} = 0  $$

En este caso las ecuaciones de movimiento son

\begin{align}\tag{2.3}

\frac{d}{ds} \left(r^2 \left(\frac{d\varphi}{ds}\right)\right) &= 0  \qquad \Longrightarrow \qquad \frac{d\varphi}{ds} = \frac{J}{r^2};\\
\frac{d}{ds} \left[\left(1 - \frac{2M}{r}\right) \left(\frac{dt}{ds}\right)\right] &= 0 \qquad \Longrightarrow \qquad \frac{dt}{ds} = \left(1 -\frac{2M}{r}\right)^{-1} E \tag{2.4}

\end{align}


Reemplazando (2.3) y (2.4) en (2.2), resulta: 

\begin{equation*}
    1 = \left( 1 - \frac{2M}{r}\right)^{-1} E^2 -\left( 1 - \frac{2M}{r}\right)^{-1} \left(\frac{dr}{ds}\right)^2 - \frac{J^2}{r^2}
\end{equation*}

para concluir que

\begin{equation}\tag{2.5}
    \left(\frac{dr}{ds}\right)^2 = E^2 -\left( 1 - \frac{2M}{r}\right) \left( 1 + \frac{J^2}{r^2}\right)
\end{equation}

Ahora, diferenciando (2.5) con respecto a $s$ obtenemos

\begin{equation}\tag{2.6}
    2\frac{d^2r}{ds^2} = \frac{d}{dr}\left(E^2 -\left( 1 - \frac{2M}{r}\right) \left( 1 + \frac{J^2}{r^2}\right)\right)
\end{equation}

Al igual que en el problema de Kepler es conveniente tratar $r$ como función de $\varphi$. Sea 

$$u(\varphi) = \frac{M}{r(s(\varphi))} \tag{2.7}$$

entonces

\begin{equation*}
\frac{du}{d\varphi} = - \frac{M}{r^2} \frac{dr}{ds} \frac{ds}{d\varphi} = - \frac{M}{r^2} \frac{\frac{dr}{ds}}{\frac{ds}{d\varphi}} = - \frac{M}{J} \frac{dr}{ds}.
\end{equation*}

Así, la ecuación de movimiento de segundo orden para $u$ puede obtenerse diferenciando (2.6) con respecto a $\varphi$, lo que lleva a 


  $$   2\frac{d^2u}{ds^2} \frac{du}{d\varphi} = \frac{d}{du}\left( -u^2(1-2u) - \frac{M^2 (1-2u)}{J^2} \right) \frac{du}{d\varphi} ;$$

  por lo que o bien $r$ es constante (orbita circular)  o

\begin{equation*}\tag{2.8}
\frac{d^2u}{ds^2}   =  -u + 3 u^2 + \frac{M^2}{J^2} 
\end{equation*}

##### Análisis del potencial efectivo

El análisis de las geodésicas en la métrica de Schwarzschild se puede simplificar al estudio del movimiento unidimensional descrito por $u = u(\varphi)$ en un potencial efectivo $V(u)$ dado por la siguiente expresión:

$$V(u) = \frac{1}{2}u^2 - u^3 - \frac{m^2}{J^2} u. $$

En el caso newtoniano, se elimina la corrección relativista lo que implica que consideramos el mismo potencial, pero sin el término $u^3$.

```python
#Potencial efectivo sin correccion relativista (caso Newtoniano)
def VN(u,m,J):
    return u**2/2 - (m/J)**2*u 

#Potencial efectivo con correccion relativista
def Vr(u,m,J):
    return u**2/2 - u**3 - (m/J)**2*u


#Parametros teniendo encuenta 0< m/J < 1/4 (Soluciones estables)
m = 1
L = 5

#Grafica del potencial efectivo
fig , axs = plt.subplots(1,2, figsize=(10,3))
ax = axs.flatten()

u = np.linspace(0, 1, 1000)
ax[0].plot(u, Vr(u,m,L), 'b', label='u(t)')
ax[0].plot(u, VN(u,m,L), 'y--', label='u(t)')
ax[0].set_xlim(0, 0.5)
ax[0].set_ylim(-0.03,0.04)
#ax[0].hlines(0, 0, 1, colors='k', linestyles='dashed')
ax[0].set_xlabel('u')
ax[0].set_ylabel('V(u)')
#ax[0].set_title('Potencial efectivo')
ax[0].grid()

ax[1].plot(u, Vr(u,m,L), 'b', label='u(t)')
ax[1].plot(u, VN(u,m,L), 'y--', label='u(t)')
ax[1].set_xlim(0, 0.12)
ax[1].set_ylim(-0.0010,0.0025)
#ax[1].hlines(0, 0, 1, colors='k', linestyles='dashed')
ax[1].set_xlabel('u')
ax[1].set_ylabel('V(u)')
#ax[1].legend()
#ax[1].set_title('Potencial efectivo')
ax[1].grid()

plt.subplots_adjust(wspace=0.5)
plt.suptitle('Potencial efectivo con correccion relativista', fontsize=15)
plt.show()
```

El gráfico de la izquierda muestra la forma general del potencial, y el gráfico discontinuo corresponde al caso newtoniano. El de la derecha es un acercamiento a la región de $u$ pequeño. Todas las órbitas de energía negativa que se encuentran en la región cubierta por el segundo gráfico permanecen allí y son periódicas en $\varphi$ con un período distinto de $2 \pi $ , excepto la órbita constante en el mínimo del potencial (órbita circular). 

##### Exploración de las órbitas mediante soluciones analíticas

Por otro lado, podemos resolver la ecuación (2.8) utilizando el método de Poincaré-Lindstedt. A continuación, presentamos la solución aproximada sin detallar su desarrollo por motivos prácticos: 

\begin{equation*}
u(\varphi)   \approx  \frac{m^2}{J^2}(1+e \cos(\varphi)) + \frac{m^4}{J^4}\left[ -(3+e^2)\cos(\varphi) + 3\left(1 + \frac{e^2}{2}\right) -\frac{e^2}{2}\cos(2\varphi) + 3e\varphi \sin(\varphi) \right]
\end{equation*}

Ahora, podemos comparar esta solución aproximada con la solución analítica (1.3) de la ecuación de la órbita newtoniana. Graficar ambas soluciones nos permitirá realizar una comparación visual.


```python
#Para este analisis tomamos A = m/J y G = m0 = 1, por simplicidad.

#Definimos la solucion analitica de la ecuacion de la orbita Newtoniana
def u(phi,e,A):
    #A = m/J
    return A**2*(1+e*np.cos(phi))

#Definimos la solucion aproximada de la ecuacion de la orbita relativista 
def u_(phi,e,A):
    #A = m/J
    return A**2*(1+e*np.cos(phi))+ A**4*(-(3+e**2)*np.cos(phi)+3*(1+e**2/2)-e**2*np.cos(2*phi)/2+3*e*phi*np.sin(phi))

#Definimos las condiciones iniciales
y = [0,0]
e = 0.7 #Excentricidad de orbitas elipticas
A = 1/15 # m=1; J=15

#Coordenadas de la orbita Newtoniana
t = np.linspace(0, 4*np.pi, 1000)

x = (1/u(t,e,A))*np.cos(t)
y = (1/u(t,e,A))*np.sin(t)
z = np.zeros_like((1/u(t,e,A)))  # Plano z=0

#Coordenadas de la orbita relativista
t_ = np.linspace(0, 8*np.pi, 1000)

x_ = (1/u_(t_,e,A))*np.cos(t_)
y_ = (1/u_(t_,e,A))*np.sin(t_)
z_ = np.zeros_like((1/u_(t_,e,A)))  # Plano z=0


#Gráfico de la orbita Newtoniana y relativista
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

ax.plot(x_, y_, z_,color = '#E92EFB', label='Orbita relativista')
ax.plot(x, y, z,color = '#FF2079', label='Orbita Newtoniana')
ax.legend()

ax.xaxis.pane.set_facecolor('#04005E')
ax.yaxis.pane.set_facecolor('#04005E')
ax.zaxis.pane.set_facecolor('#04005E')
ax.set_facecolor('#04005E')

ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')
ax.tick_params(axis='z', colors='white')

ax.set_xlabel("x", color='white')
ax.set_ylabel("y", color='white')
ax.set_zlabel("z", color='white')

#cambiamos el angulo de vista
ax.view_init(15, -66)

#plt.title("Orbita Newtoniana y relativista", fontsize=15)
plt.show()
```

En el gráfico se puede observar un incremento en el eje mayor de la órbita relativista, lo cual es consistente con la solución aproximada que contiene términos seculares que generan resonancia. Además, se pueden identificar dos puntos en los que la órbita relativista pasa repetidamente. Como no es el resultado que queremos, podemos reemplazar el término secular $\varphi \sin(\varphi)$ por el período modificado de una aproximación Kepleriana a la órbita de Schwarzschild, obtenida mediante el método de Poincaré-Lindstedt. Sin embargo, en este caso, simplemente resolveremos numéricamente la ecuación (2.8).

##### Exploración de las órbitas mediante soluciones numéricas

```python
#Ecuaion de la orbita Newtoniana
def u_N(y,phi,A):
    u , u_dot = y
    return [u_dot, -u + A**2]

#Ecuaion de la orbita Relativista
def u_R(y,phi,A):
    u , u_dot = y
    return [u_dot, -u + 3*u**2 + A**2]

#Solucion de la ecuacion de la orbita Newtoniana con Odeint
y_ = [0.0005,0] #condiciones iniciales
A = 1/15 # m=1; J=15

phi_ = np.linspace(0, 4*np.pi, 2000)

#Solucion de la ecuacion de la orbita Newtoniana con Odeint
sol_ = odeint(u_N, y_, phi_, args=(A,))

#coordenadas de la orbita Newtoniana
x_ = (1/sol_[:,0])*np.cos(phi_)
y_ = (1/sol_[:,0])*np.sin(phi_)
z_ = np.zeros_like((1/sol_[:,0]))  # Plano z=0

#Solucion de la ecuacion de la orbita Relativista con Odeint

y = [0.0005,0] #condiciones iniciales

m = 1
phi = np.linspace(0, 10*np.pi, 2000)

#Solucion de la ecuacion de la orbita Relativista con Odeint
sol = odeint(u_R, y, phi, args=(A,))

#coordenadas de la orbita Relativista
x = (m/sol[:,0])*np.cos(phi)
y = (m/sol[:,0])*np.sin(phi)
z = np.zeros_like((m/sol[:,0]))  # Plano z=0

# Crear figura y ejes 3D
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, z,color = '#E92EFB', label='Orbita relativista')
ax.plot(x_, y_, z_,color = '#FF2079', label='Orbita Newtoniana' , linewidth=2)
ax.legend()

ax.xaxis.pane.set_facecolor('#04005E')
ax.yaxis.pane.set_facecolor('#04005E')
ax.zaxis.pane.set_facecolor('#04005E')
ax.set_facecolor('#04005E')

ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')
ax.tick_params(axis='z', colors='white')

ax.set_xlabel("x", color='white')
ax.set_ylabel("y", color='white')
ax.set_zlabel("z", color='white')

#cambiamos el angulo de vista
ax.view_init(20, -60)
```

 De la grafica podemos notar que los ejes mayores de las elipses generadas por la órbita relativista no coinciden entre sí. Esto implica que la posición del perihelio, es decir, el punto más cercano al objeto central, experimenta pequeños desplazamientos. Este fenómeno es conocido como desplazamiento del perihelio. Además, es importante destacar que mediante el análisis de la órbita newtoniana no se evidencia este tipo de fenómeno. 

En conclusión, al utilizar la métrica de Schwarzschild y aplicar métodos numéricos, hemos logrado demostrar que las geodésicas en el espacio-tiempo de Schwarzschild concuerdan con las observaciones existentes sobre la órbita de Mercurio. La elección de Mercurio como referencia se debe a que presenta un desplazamiento en su perihelio, un fenómeno que no puede ser explicado con precisión mediante la mecánica newtoniana, como hemos demostrado.
